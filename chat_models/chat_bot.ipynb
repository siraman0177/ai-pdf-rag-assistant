```python
from dotenv import load_dotenv
import os
from langchain_mistralai import ChatMistralAI

load_dotenv(override=True)

model = ChatMistralAI(
    model="mistral-small-latest", 
    temperature=0.7,
    max_tokens=100
)


messages=[]
print("---------type 0 to exit-------------")
while True:
    
    prompt=input("you : ")
    messages.append(prompt)
    if prompt=="0": break 
    response = model.invoke(prompt)
    messages.append(response.content)
    print("bot : ",response.content)
```

```python
from dotenv import load_dotenv
import os
from langchain_mistralai import ChatMistralAI
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage

load_dotenv(override=True)

model = ChatMistralAI(
    model="mistral-small-2506", 
    temperature=0.9,
    max_tokens=100
)
print(" choose mode of the AI")
print(" choose 1 for angry")
print(" choose 2 for sad")
print(" choose 3 for happy")

choose=int(input("tell your respose"))
if choose == 1 : 
    mode= "you are a angry agent and reply everything angrly "
elif choose ==2 :
    mode="you are a sad agent and eply every thing in sad way"
else : mode="You are a funny AI agent who replies humorously."

messages=[
    SystemMessage(content= mode)
]
print("---------type 0 to exit-------------")
while True:
    
    prompt=input("you : ")
    
    messages.append(HumanMessage(content=prompt))
    if prompt=="0": break 
    response = model.invoke(messages)
    messages.append(AIMessage(content=response.content))
    print("bot : ",response.content)
```
