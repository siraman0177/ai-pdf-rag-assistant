```python
from dotenv import load_dotenv
from langchain_mistralai import ChatMistralAI
from langchain_community.document_loaders import TextLoader
from langchain_core.prompts import ChatPromptTemplate

data=TextLoader("note.txt")
docs=data.load()
template=ChatPromptTemplate.from_messages([
    ("system", "you summarize the text"),
    ("human", "{text}")
])
model =ChatMistralAI(model= "mistral-small-latest")
prompt=template.format_prompt(text=docs[0].page_content)
result = model.invoke(prompt)
print(result.content)
```

# Retrivers

```python
from langchain_community.retrievers import ArxivRetriever

#create retriever
retriever= ArxivRetriever(
    load_max_docs=2,         #number of paper to retrive 
    load_all_available_meta=True
)

# query 
docs= retriever.invoke("Generative AI")

#print 
for i, doc in enumerate(docs):
    print(f"\n Result {i+1}:")
    print(f"Title: {doc.metadata['title']}")
    print(f"Authors: {', '.join(doc.metadata['authors'])}")
    print(f"Abstract: {doc.page_content}")
    print("--------------------------")
```

```python
import arxiv

search = arxiv.Search(
    query="Generative AI",
    max_results=2
)

for paper in search.results():
    print("Title:", paper.title)
    print("Authors:", [a.name for a in paper.authors])
    print("Summary:", paper.summary)
    print("-"*50)
```
